```
file_path = input("file path plz :)(magazine.json) >>")
magazine_data = pd.read_json(file_path, lines=True)

magazine_data.head(5)
```

* magazine.json: 매거진 정보

  * 총 27,967개의 브런치 매거진 정보입니다.
  * 필드 설명
    * id: 매거진 식별자
    * magazine_tag_list: 작가가 부여한 매거진의 태그 정보


# magazine_tag_list

```
magazine_tag_list = pd.DataFrame(Counter([j for i in magazine_data["magazine_tag_list"].tolist() for j in i]).most_common())

magazine_tag_list.head(5)
```
